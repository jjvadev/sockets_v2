```python
import json
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

csv_file = Path(r"./results\cifar100_w1_e50_le1_bs128_lr0.01_leaky_relu\history.csv")
json_file = Path(r"./results\cifar100_w1_e50_le1_bs128_lr0.01_leaky_relu\summary.json")

df = pd.read_csv(csv_file)
with open(json_file, 'r', encoding='utf-8') as f:
    summary = json.load(f)

df.head()
```

## Curvas principales


```python
plt.figure(figsize=(8,5))
plt.plot(df['epoch'], df['train_cost'], marker='o', label='Costo Train')
plt.plot(df['epoch'], df['test_cost'], marker='s', label='Costo Test')
plt.title('Costo por época')
plt.xlabel('Época')
plt.ylabel('Costo')
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(8,5))
plt.plot(df['epoch'], df['train_acc'], marker='o', label='Train Acc')
plt.plot(df['epoch'], df['test_acc'], marker='s', label='Test Acc')
plt.title('Accuracy por época')
plt.xlabel('Época')
plt.ylabel('Accuracy')
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(8,5))
plt.plot(df['epoch'], df['epoch_time'], marker='o')
plt.title('Tiempo por época')
plt.xlabel('Época')
plt.ylabel('Segundos')
plt.grid(True)
plt.show()

plt.figure(figsize=(8,5))
plt.plot(df['epoch'], df['total_time'], marker='o')
plt.title('Tiempo acumulado')
plt.xlabel('Época')
plt.ylabel('Segundos')
plt.grid(True)
plt.show()

plt.figure(figsize=(8,5))
plt.plot(df['epoch'], df['train_acc'] - df['test_acc'], marker='o')
plt.title('Gap Train-Test')
plt.xlabel('Época')
plt.ylabel('Gap')
plt.grid(True)
plt.show()
```
